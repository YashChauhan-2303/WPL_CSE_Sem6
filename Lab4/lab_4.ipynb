Q1.Write a python program to reverse a content a file and store it in another file.

```python
with open("input.txt", "r") as f1:
    data = f1.read()

with open("output.txt", "w") as f2:
    f2.write(data[::-1])

print("File reversed successfully.")
```

Q2. Write a python program to implement binary search with recursion.

```python
def binary_search(arr, low, high, key):
    if low <= high:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] > key:
            return binary_search(arr, low, mid - 1, key)
        else:
            return binary_search(arr, mid + 1, high, key)
    return -1

arr = [10,20,30,40,50,60]
key = 40
result = binary_search(arr, 0, len(arr)-1, key)

print("Found at index:", result if result != -1 else "Not Found")
```

Q3. Write a python program to sort words in alphabetical order.

```python
s = input("Enter words: ")
words = s.replace(',', '').split()
words.sort(key=str.lower)
print("Sorted words:", words)
```

Q4. Write a Python class to get all possible unique subsets from a set of distinct
integers Input:[4,5,6]
Output : [[], [6], [5], [5, 6], [4], [4, 6], [4, 5], [4, 5, 6]]

```python
from itertools import combinations

nums = [4,5,6]
subsets = []

for i in range(len(nums)+1):
    for comb in combinations(nums, i):
        subsets.append(list(comb))

print(subsets)
```

Q5. Write a Python class to find a pair of elements (indices of the two numbers)
from a given array whose sum equals a specific target number.
Input: numbers= [10,20,10,40,50,60,70], target=50
Output: 3, 4.

```python
class Solution:
    def twoSum(self, nums, target):
        for i in range(len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    return i, j

nums = [10,20,10,40,50,60,70]
target = 50

print(Solution().twoSum(nums, target))
```

Q6. Write a Python class to implement pow(x, n).

```python
class Power:
    def pow(self, x, n):
        return x ** n

p = Power()
print(p.pow(2,5))
```

Q7. Write a Python class which has two methods get_String and print_String. The
get_String accept a string from the user and print_String print the string in upper
case.

```python
class String:
    def get_String(self):
        self.s = input("Enter string: ")

    def print_String(self):
        print(self.s.upper())

obj = String()
obj.get_String()
obj.print_String()
```
